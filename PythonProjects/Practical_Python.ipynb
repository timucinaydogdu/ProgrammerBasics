# Projelerle Egitim

#### Temel Veri Tipleri

- Integer   ;   Tam sayi veri tipi 
- Float     ;   Ondalik sayi veri tipi 
- Boolean   ;   Mantiksal veri tipi (True/False)
- String    ;   Alfanumerik veri tipi (''/"")

```python
# Yorum satiridir. Altta veri tipleri kullanimi ve type kullanimi bulunmaktafir.
print(15,type(15))
print(85.12,type(85.12))
print(True,type(True))
print("Nehir",type("Nehir"))
```

```python
# Degiskenler kendilerine deger atanan degerlerdir. Tekrar tekrar kullanilabilir.
x = 15
y = 85.12
z = True
t = "Nehir"

print(x,type(x))
print(y,type(y))
print(z,type(z))
print(t,type(t))
```

```python
# Print icinde fazla degisken ve veri kullanimi

yas = 15
ortalama = 85.2
durum = True
ad = "Nehir"
print("Benim adim",ad,"yasim",yas,"ortalamam ",ortalama,"okula devam durumu", durum)

print("=======================================================================")

# Print degisken enjecksiyonu

print(f"Benim adim {ad} yasim {yas} ortalamam {ortalama} okula devam durumu : {durum}")
```

```python
# Aritmetik Operatorler

print(5+2)
print(5-2)
print(5*2)
print(5/2)  # sonuc herzaman float deger olusturur.

print("=======================================================================")

print(5**2) # ust almak icin kullanilir.
print(5%2)  # mod almak icin kullanilir.
print(5//2) # 5 degerinde kac tane 2 var.

print("=======================================================================")

# Strinlerde aritmetiksel operatorler.
ad = "Nehir"
soyad = "uzun"

print(ad + soyad)
print(ad * 5)
print("=======================================================================")

# Karsilastirma operatorleri. Sonuclar true yada false olarak verir.
yas = 15

print(5==9,"=5==9")          # esit mi ?
print(5==5,"=5==5")
print(5!=9,"=5!=9")          # esit degil mi ?
print(5!=5,"=5!=5")

print(yas>=5,"=yas>=5")      # buyuk esit mi ?
print(yas>5,"=yas>5")        # buyuk mu ?
print(yas<5,"=yas<5")        # kucuk mu ?
print(yas<=5,"=yas<=5")      # kucuk esit mi ?


```

```python
# Mantiksal operatorlerin kullanimi. True ve false doner.
print("And Kullanimi")
print(True and True,"=true and true")
print(True and False,"=true and false")
print(False and True,"=false and true")
print(False and False,"=false and false")

print("Or Kullanimi")
print(True or True,"=true or true")
print(True or False,"=true or false")
print(False or True,"=false or true")
print(False or False,"=false or false")

print("=======================================================================")

print(not True,"=not true")
print(not False,"=not false")
```

```python
veri = "35"
sayi = 25

print(type(veri), veri, veri*2) # Dikkat edilmesi gereken nokta.
print(type(sayi), sayi, sayi*2)

print("=======================================================================")

intveri = int(veri)
strsayi = str(sayi)

print(type(intveri), intveri, intveri*2) # string bir degerin integer bir degere donusumu.
print(type(strsayi), strsayi, strsayi*2) # ingeri bir sayinin string bir degere  donusumu.
```

```python
puan1 = float(input("Birinci notunuzu giriniz."))
puan2 = float(input("Ikinci notunuzu giriniz."))

devamsizlik = float(input("Devamsizlik sayisi"))
ortalama = (puan1 + puan2) / 2
gecmeDurumu = (ortalama >= 50 and devamsizlik <= 10)

print("Ortalama ", ortalama)
print("Gecme Durumu ",gecmeDurumu)
print(f"Birinci puan {puan1} ve ikinci {puan2} dir. Ortalama {ortalama} dir. Devamsizlik {devamsizlik} gundur. Ondan dolayu {gecmeDurumu}")
```

#### Kosul yapilari (IF)

```python
ad = input("Adinizi giriniz.")
if ad == "nehir":
    print(f"Adiniz {ad} dir.")
```

```python
ortalama = float(input("Ortalamanizi giriniz."))
if ortalama >= 50 :     # True durumu icin kullanilan kisim.
    print(f"Ortalamaniz {ortalama} dir. Gectiniz.")
else:                   #  False durumu icin kullanilan kisim.
    print("Ortalmaniz gecmenize yetmiyor.")
```

```python
ortalama = float(input("Ortalamanizi giriniz."))
devamsizlik = float(input("Devamsizlik sayisini giriniz."))
if ortalama >= 50 :

    print(f"Ortalamaniz {ortalama} dir. Gectiniz.")
    if devamsizlik <= 10 :   # Kosul icinde ayri bir kontrol olarak kullanilir.
        print("Devamsizlik durumu uygun.")
    else:
        print("Devamsizlik durumu uygun degil.")

else:
    print("Ortalmaniz gecmenize yetmiyor.")
```

```python
ortalama = float(input("Ortalamanizi giriniz."))
devamsizlik = float(input("Devamsizlik sayisini giriniz."))
if ortalama < 50 :     # Negatif kosul
    print(f"Ortalamaniz {ortalama} dir. Kaldiniz.")
elif devamsizlik >= 10: # Negatif secenekli kosul
    print("Devamsizlik durumu uygun degil.")
else:                   # Pozitif kosul
    print("Sinifinizi gectiniz.")
```

```python
sinif = int(input("Sinifinizi giriniz."))
if sinif <= 4 :
    print("Ilk okul")
elif sinif <= 8 :
    print("Orta okul")
elif sinif <= 12 :
    print("Lise")
else:
    print("Universite")
```

```python
ortalama = float(input("Ortalamanizi girin."))
print("Gecti") if ortalama >= 50 else print("KALDI..") # Tek satirda if kontrolu.
```

#### Kosul yapilari (Match CASE); liste bazli bir secim dallanmasidir.

```python
bolum_kodu = int(input("Bolum kodunu giriniz."))
match bolum_kodu :
    case 1:
        print("Bilisim")
    case 2:
        print("Elektrik")
    case 3:
        print("Mobilya")
```
