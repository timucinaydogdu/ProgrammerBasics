```python
print("-" * 40)
# Basic printing examples
print("Hello, World!")
print("Welcome to Python programming.")
print("This is a simple script to demonstrate printing in Python.")
print("You can print multiple lines of text using the print function.")
print("-" * 40)

# Data types demonstration
print("Integer example:", 42)
print("Float example:", 3.14159)
print("String example:", "Python is fun!")
print("Boolean example:", True)
print("-" * 40)

# Demonstrating various arithmetic operations
x = 10
y = 20

# Arithmetic comparison examples
print("Is x equal to y?", x == y)
print("Is x not equal to y?", x != y)
print("Is x greater than y?", x > y)
print("Is x less than y?", x < y)
print("Is x greater than or equal to y?", x >= y)
print("Is x less than or equal to y?", x <= y)  
print("-" * 40)

# Arithmetic operations examples
print("The value of x is:", x)
print("The value of y is:", y)
print("The sum of x and y is:", x + y)
print("The difference of x and y is:", x - y)
print("The product of x and y is:", x * y)
print("The quotient of x and y is:", x / y)
print("The integer division of x and y is:", x // y)
print("The remainder of x divided by y is:", x % y)
print("The exponentiation of x to the power of y is:", x ** y)
print("-" * 40)
```

```python
print("-" * 40)
# Using built-in functions
print("Rounded value of 3.14159 to 2 decimal places:", round(3.14159, 2))
print("The absolute value of -42 is:", abs(-42))
print("The maximum of x and y is:", max(x, y))
print("The minimum of x and y is:", min(x, y))
print("-" * 40)

# String manipulation examples
TextDisplayObject = "This is a text display object."
print(TextDisplayObject)
print("This is a divided Text: " + TextDisplayObject[3:15])
print("Length of TextDisplayObject:", len(TextDisplayObject))
print("Count of 't' in TextDisplayObject:", TextDisplayObject.count('t'))
print("Uppercase:", TextDisplayObject.upper())
print("Lowercase:", TextDisplayObject.lower())
print("-" * 40)

# Concatenating strings
text1 = "Hello"
text2 = "World"
combined_text = text1 + ", " + text2 + "!"
print("Combined text: " + combined_text)
print("-" * 40)


# Using f-strings for formatted output
print(f"Text1: {text1}, Text2: {text2}, CombinedText: {combined_text}")
print("-" * 40)

# string casting
number = 42
print("The number is: " + str(type(number)) + " now " + str(type(str(number))))
print("-" * 40)
```

```python
print("-" * 40)
# List examples
my_list = [1, 2, 3, 4, 5]
my_list.append(6)
print("List after appending 6:", my_list)
my_list.remove(3)
print("List after removing 3:", my_list)
print("Length of the list:", len(my_list))
print(count := my_list.count(2), "occurrences of 2 in the list")
print("Reversed list:", my_list[::-1])
print("Original list:", my_list)
print("First element:", my_list[0])
print("Last element:", my_list[-1]) 
print("Sliced list (elements 1 to 3):", my_list[1:4])
print("List after multiplying by 2:", my_list * 2)
print("List after concatenation:", my_list + [6, 7, 8])
print("Is 3 in the list?", 3 in my_list)
print("Is 10 not in the list?", 10 not in my_list)
print("Index of element 4 in the list:", my_list.index(4))
print("List after sorting:", sorted(my_list, reverse=True))
print("List after sorting in ascending order:", sorted(my_list))
print("Maximum value in the list:", max(my_list))
print("Minimum value in the list:", min(my_list))   
print("Sum of all elements in the list:", sum(my_list))
print("Average of elements in the list:", sum(my_list) / len(my_list))
print("List after multiplying each element by 2:", [x * 2 for x in my_list])
print("List after filtering even numbers:", [x for x in my_list if x % 2 == 0])
print("List after squaring each element:", [x ** 2 for x in my_list])
print("List after converting each element to string:", [str(x) for x in my_list])
print("List after converting each element to float:", [float(x) for x in my_list])
print("Sum of squares of elements in the list:", sum(x ** 2 for x in my_list))
print("List after removing duplicates:", list(set(my_list)))
print("List after sorting unique elements:", sorted(set(my_list)))
print("List after reversing using reverse():", end=" ")
my_list.reverse()
print(my_list)
print("-" * 40)
```

```python
print("-" * 40)
# Tuple examples : tuples are immutable
my_tuple = (1, 2, 3, 4, 5)
print("Original tuple:", my_tuple)
print("Length of the tuple:", len(my_tuple))
print("First element:", my_tuple[0])
print("Last element:", my_tuple[-1])
print("Sliced tuple (elements 1 to 3):", my_tuple[1:4])
print("Tuple after concatenation:", my_tuple + (6, 7, 8))
print("Is 3 in the tuple?", 3 in my_tuple)
print("Is 10 not in the tuple?", 10 not in my_tuple)
print("Index of element 4 in the tuple:", my_tuple.index(4))
print("Count of element 2 in the tuple:", my_tuple.count(2))
# Tuples are immutable, so we cannot modify them directly
# However, we can create a new tuple by concatenating or slicing
new_tuple = my_tuple[:2] + (99,) + my_tuple[3:]
print("New tuple after modification:", new_tuple)
print("-" * 40)
```

```python
print("-" * 40)
# Sets examples : their unordered and unindexed nature
my_set = {1, 2, 3, 4, 5}
print("Original set:", my_set)
my_set.add(6)
print("Set after adding 6:", my_set)
my_set.remove(3)
print("Set after removing 3:", my_set)
print("Length of the set:", len(my_set))
print("Is 4 in the set?", 4 in my_set)
print("Is 10 not in the set?", 10 not in my_set)
another_set = {4, 5, 6, 7, 8}
print("Another set:", another_set)
print("Union of sets:", my_set.union(another_set))
print("Intersection of sets:", my_set.intersection(another_set))
print("Difference of sets (my_set - another_set):", my_set.difference(another_set))
print("Symmetric difference of sets:", my_set.symmetric_difference(another_set))
print("-" * 40)
```

```python
print("-" * 40)
# Dictionary examples
my_dict = {'a': 1, 'b': 2, 'c': 3}
print("Original dictionary:", my_dict)
my_dict['d'] = 4
print("Dictionary after adding key 'd':", my_dict)
del my_dict['b']
print("Dictionary after deleting key 'b':", my_dict)
my_dict.update({'e': 5})
print("Updated dictionary:", my_dict)   
print("Dictionary after clearing all items:", end=" ")
temp_dict = my_dict.copy()
temp_dict.clear()
print(temp_dict)  
print("Length of the dictionary:", len(my_dict))
print("Value for key 'a':", my_dict['a'])
print("Is key 'c' in the dictionary?", 'c' in my_dict)
print("Is key 'b' not in the dictionary?", 'b' not in my_dict)
print("Dictionary keys:", my_dict.keys())
print("Dictionary values:", my_dict.values())
print("Dictionary items:", my_dict.items())
print("Get value for key 'c':", my_dict.get('c'))
print("Get value for non-existing key 'z' with default:", my_dict.get('z', 'Not Found'))
print("-" * 40)
```

```python
print("-" * 40)
# conditionals and boolean logic
# Comparison operators: ==, !=, >, <, >=, <=
# Logical operators: and, or, not

a = 10
b = 20  
print("a =", a)
print("b =", b) 

if a < b:
    print("a is less than b")
elif a == b:
    print("a is equal to b")
else:
    print("a is greater than b")
print("-" * 40)     

if a < b and b > 15:
    print("Both conditions are true")
if a < b or b < 15:
    print("At least one condition is true")
if not a > b:
    print("a is not greater than b")

print('a id:', id(a), 'b id:', id(b))
print('Are a and b the same object?', id(a) == id(b))

print("-" * 40) 

```

```python
print("-" * 40)
# Loops and iterations

# For loop example
print("For loop example:")
for i in range(5):
    print("Iteration:", i)
print("-" * 40) 

# While loop example
print("While loop example:")
count = 0
while count < 5:
    print("Count:", count)
    count += 1
print("-" * 40) 

# Looping through a list
fruits = ["apple", "banana", "cherry"]
print("Looping through a list:")
for fruit in fruits:
    print(fruit)    
print("-" * 40) 

# Looping through a dictionary
student = {"name": "Alice", "age": 20, "grade": "A"}
print("Looping through a dictionary:")
for key, value in student.items():
    print(f"{key}: {value}")
print("-" * 40)

# Using break and continue
print("Using break and continue:")
for i in range(10):
    if i == 5:
        break  # Stop the loop when i is 5
    if i % 2 == 0:
        continue  # Skip even numbers
    print(i)
print("-" * 40)
```

```python
print("-" * 40)

# Functions
def helloFunction():
    print("Hello from the function!")   
helloFunction()
print("-" * 40)

def addNumbers(a, b):
    return a + b    
result = addNumbers(5, 10)
print("Sum of 5 and 10 is:", result)
print("-" * 40)

def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
num = 5
print(f"Factorial of {num} is:", factorial(num))
print("-" * 40)

def isEven(n):
    return n % 2 == 0
number = 4
print(f"Is {number} even?", isEven(number))     
print("-" * 40)

def studentinfo(*args, **kwargs):
    print("Positional arguments:", args)
    print("Keyword arguments:", kwargs)
studentinfo("Alice", 20, grade="A", major="Computer Science")
print("-" * 40)
```
