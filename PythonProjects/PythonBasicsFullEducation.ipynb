```python
# Hello World program in Python
print("Hello, World!")

# Arithmetic operations
print("Addition: " + str(2 + 3))             # addition(+)
print("Subtraction: " + str(3 - 1))          # subtraction(-)
print("Multiplication: " + str(2 * 3))       # multiplication(*)
print("Division: " + str(3 / 2))             # division(/)
print("Exponential: " + str(3 ** 2))         # exponential(**)
print("Modulus: " + str(3 % 2))              # modulus(%)
print("Floor Division: " + str(3 // 2))      # Floor division operator(//)

# Checking data types
print(type(10))          # Int
print(type(3.14))        # Float
print(type(1 + 3j))      # Complex number
print(type('Asabeneh'))  # String
print(type([1, 2, 3]))   # List
print(type({'name':'Asabeneh'})) # Dictionary
print(type({9.8, 3.14, 2.7}))    # Set
print(type((9.8, 3.14, 2.7)))    # Tuple
```
