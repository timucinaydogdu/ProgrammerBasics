```python
# Different python data types
first_name = 'Asabeneh'                                 # str
last_name = 'Yetayeh'                                   # str
country = 'Finland'                                     # str
city= 'Helsinki'                                        # str
age = 250                                               # int, it is not my real age, don't worry about it

# Printing out types
print("first_name: Asabeneh, type: ", type(first_name)) # str
print("last_name: Yetayeh, type: ", type(last_name))    # str
print("country: Finland, type: ", type(country))        # str
print("city: Helsinki, type: ", type(city))             # str
print("age: 250, type: ", type(age))                    # int
print("3.14, type: ", type(3.14))                       # float
print("1 + 3j, type: ", type(1 + 3j))                   # complex number
print("True, type: ", type(True))                       # bool
print("False, type: ", type(False))                     # bool
print("None, type: ", type(None))                       
print("-" * 40)

# int to float
num_int = 10
print('num_int',num_int)                        # 10
num_float = float(num_int)
print('num_float:', num_float)                  # 10.0

# float to int
gravity = 9.81
print("gravity as int: " + str(int(gravity)))   # 9

# int to str
num_int = 10
print("num_int as string: " + str(num_int))     # 10
num_str = str(num_int)
print("num_str: " + num_str)                    # '10'
print("-" * 40)

# str to int or float
num_str = '10.6'
num_float = float(num_str)                      # Convert the string to a float first
num_int = int(num_float)                        # Then convert the float to an integer
print('num_int', int(num_float))                # 10
print('num_float', float(num_float))            # 10.6
num_int = int(num_float)
print('num_int', int(num_int))                  # 10

# str to list
first_name = 'Asabeneh'
print(first_name)                               # 'Asabeneh'
first_name_to_list = list(first_name)
print(first_name_to_list)                       # ['A', 's', 'a', 'b', 'e', 'n', 'e', 'h']
print("-" * 40)
```

```python
# Arithmetic operations

num_one = 3
num_two = 4

# Arithmetic operations
total = num_one + num_two
diff = num_two - num_one
product = num_one * num_two
div = num_two / num_one
remainder = num_two % num_one

# Printing values with label
print('total: ', total)
print('difference: ', diff)
print('product: ', product)
print('division: ', div)
print('remainder: ', remainder)
print("-" * 40)

# Calculating area of a circle
radius = 10                                 # radius of a circle
area_of_circle = 3.14 * radius ** 2         # two * sign means exponent or power
print('Area of a circle:', area_of_circle)
print("-" * 40)

# Calculating area of a rectangle
length = 10
width = 20
area_of_rectangle = length * width
print('Area of rectangle:', area_of_rectangle)
print("-" * 40)

# Calculating a weight of an object
mass = 75
gravity = 9.81
weight = mass * gravity
print(weight, 'N')                         # Adding unit to the weight
print("-" * 40)

# Calculate the density of a liquid
mass = 75 # in Kg
volume = 0.075 # in cubic meter
density = mass / volume # 1000 Kg/m^3
print(density, 'Kg/m^3') # Adding unit to the density
print("-" * 40)
```

```python
# Comparison operators
print("3>2: " + str(3 > 2))     # True, because 3 is greater than 2
print("3 >= 2: " + str(3 >= 2))    # True, because 3 is greater than 2
print("3 < 2: " + str(3 < 2))     # False,  because 3 is greater than 2
print("2 < 3: " + str(2 < 3))     # True, because 2 is less than 3
print("2 <= 3: " + str(2 <= 3))    # True, because 2 is less than 3
print("3 == 2: " + str(3 == 2))    # False, because 3 is not equal to 2
print("3 != 2: " + str(3 != 2))    # True, because 3 is not equal to 2
print("-" * 40)
print("len('mango') == len('avocado'): " + str(len('mango') == len('avocado')))  # False
print("len('mango') != len('avocado'): " + str(len('mango') != len('avocado')))  # True
print("len('mango') < len('avocado'): " + str(len('mango') < len('avocado')))   # True
print("len('milk') != len('meat'): " + str(len('milk') != len('meat')))      # False
print("len('milk') == len('meat'): " + str(len('milk') == len('meat')))      # True
print("len('tomato') == len('potato'): " + str(len('tomato') == len('potato')))  # True
print("len('python') > len('dragon'): " + str(len('python') > len('dragon')))   # False

# Comparing is operators
print('True == True: ', True == True)
print('True == False: ', True == False)
print('False == False:', False == False)
print("1 is 1: ", 1 == 1)           # True, because the value of 1 is 1
print("1 is 2: ", 1 == 2)           # False, because the value of 1 is not 2
print("1 is not 2: ", 1 != 2)       # True, because the value of 1 is not 2
print("-" * 40)

print(3 < 2 and 4 < 3) # False - because both statements are false
print('True and True: ', True and True)
print(3 > 2 or 4 < 3)  # True - because one of the statements is true
print('True or False:', True or False)
print(not 3 > 2)     # False - because 3 > 2 is true, then not True gives False
print(not True)      # False - Negation, the not operator turns true to false
print("-" * 40)

```

```python
# String concatenation
first_name = 'Asabeneh'
last_name = 'Yetayeh'
full_name = first_name + ' ' + last_name
print(full_name)  
print("-" * 40) 

# string multiline  
multiline_string = '''I am a teacher and I love teaching.
I teach HTML, CSS, JS, React, Python.'''
print(multiline_string) 
print("-" * 40)

# Escape characters
print('I hope everyone is enjoying the Python Challenge.\nDo you?') # New line
print('Days\tTopics\tExercises') # Tab
print('Day 1\t3\t5')
print('This is a backslash symbol (\\)') # To write a backslash
print('In every programming language it starts with "Hello, World!"') # To write double quotes inside a string we use double quotes
print('The saying \'Seeing is Believing\' isn\'t correct in 2020') # To write single quotes inside a string we use double quotes
print("-" * 40)

# String formatting old way
first_name = 'Asabeneh'
last_name = 'Yetayeh'
full_name = '{} {}'.format(first_name, last_name)
print(full_name)
print("-" * 40) 
# String formatting new way (f-string)
first_name = 'Sezai'
last_name = 'Yildiz'
full_name = f'{first_name} {last_name}'
print(full_name)
print("-" * 40) 
# String formatting with f-string and expressions
radius = 10
area = f'The area of a circle with radius {radius} is {3.14 * radius ** 2}'
print(area)
print("-" * 40) 
```

```python
# String indexing and slicing
first_name = 'Asabeneh'
print(first_name[0])  # A
print(first_name[1])  # s
print(first_name[2])  # a       
print(first_name[-6]) # a
print(first_name[-7]) # s
print(first_name[-8]) # A
print("-" * 40)
print(first_name[0:5])  # Asabe
print(first_name[0:6])  # Asaben
print(first_name[0:])   # Asabeneh
print(first_name[:])    # Asabeneh
print(first_name[0:8])  # Asabeneh
print(first_name[0:9])  # Asabeneh
print(first_name[1:])   # sabeneh
print(first_name[1:7])  # sabene
print(first_name[1:8])  # sabeneh
print(first_name[1:9])  # sabeneh
print(first_name[-8:-1]) # Asabene
print(first_name[-8:])   # Asabeneh
print(first_name[-7:])   # sabeneh
print(first_name[-6:])   # abeneh
print(first_name[-5:])   # beneh
print(first_name[-4:])   # eneh
print(first_name[-3:])   # neh
print(first_name[-2:])   # eh   
print(first_name[-1:])   # h
print("-" * 40) 
```

```python
# String Methods
first_name = 'Asabeneh'
print(first_name.upper())  # ASABENEH
print(first_name.lower())  # asabeneh
print(first_name.title())  # Asabeneh
print(first_name.capitalize())  # Asabeneh
print(first_name.swapcase())  # aSABENEH
print(first_name.count('a'))  # 2
print(first_name.find('a'))  # 1
print(first_name.index('a'))  # 1
print(first_name.isalpha())  # True
print(first_name.isdigit())  # False
print(first_name.isalnum())  # True
print(first_name.islower())  # False
print(first_name.isupper())  # False
print(first_name.isdecimal())  # False
print(first_name.isnumeric())  # False
print(first_name.isidentifier())  # False
print(first_name.isprintable())  # True
print(first_name.isascii())  # True
print(first_name.isspace())  # False
print(first_name.startswith('A'))  # True
print(first_name.endswith('h'))  # True
print(first_name.split())  # ['Asabeneh']
print(first_name.split('a'))  # ['As', 'beneh']
print(first_name.replace('a', 'x'))  # Asxbeneh
print("-" * 40)
```
