````python
def unique(x):
    
    dynamic_list = x.copy()
    unique_list = []
    
    for i in x:
        if dynamic_list.count(i) > 1:
            dynamic_list.remove(i)
            
        else:
            unique_list.append(i)
    return unique_list

sample_list = [1,1,3,3,3,3,4,4,4,5,7,7,7]

print(unique(sample_list))
````

4- Bir sayÄ± parametresi alan ve sayÄ±nÄ±n asal olup olmadÄ±ÄŸÄ±nÄ± kontrol eden bir Python fonksiyonu yazÄ±n.

Ä°pucu:  
Asal sayÄ±, 1â€™den bÃ¼yÃ¼k ve kendisi ile 1 dÄ±ÅŸÄ±nda pozitif bÃ¶leni olmayan doÄŸal sayÄ±dÄ±r.


````python
def is_prime(x):
    is_prime = True

    if type(x) == float or x <= 1:
        is_prime = "Un-Defined"    
    else:
        for i in range(2,x):
            if x%i == 0:
                is_prime = False
                break
                
    return is_prime
````

````python
print(is_prime(5.3))
````
