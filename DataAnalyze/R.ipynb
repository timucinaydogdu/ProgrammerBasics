* Hazir Fonksiyonlar islemleri

```r
ls()            # Enviroentmenteki nesnelerin adlarini verir.  

rm(a)           # Degiskenlerden belirtilen veriyi silmeye yarar. 

round(3.1415)   # Tam sayi yuvarlama. 

factorial(5)    # Factoriyel hesaplama. 

mean(b)         # Verilen dizenin ortalama hesabi icin kullanilir. 

sample(x,size)  # Verilen x araliginda, size adedince ornek almak icin kullanirlir. 

sample(
  x = 1:4, 
  size = 2
  ) # Random fonksiyonuna benzer. Belirli aralikta belirli adette ornek uretir. x degeri alacagi ornek araligini belirtirken size ise ornek miktaridir. 

zar <- 1:6
sample(x=zar,size = 3)
    # Zar ornegi, 3 zar atisi degeri istenmistir.

args()          # isleme alinan forksiyonun kullabilir argumanlarini verir.
args(sample)    # Sample fonksiyonun x, size gibi kullanimlarini verir. 
```
* Fonksiyonlar islemleri

Olusum sablonuna dikkat edilmeli. Fonksiyon deger atama ile bitiyorsa deger dondurmez. <br>
Herhangi bir islemle veya cagirimla bitiyorsa deger dondurur.<br>

```r
roll <- function(){
  zar <- 1:6
  zarat <- sample(
    x=zar,
    size = 2,
    replace = TRUE  
        # Replace komutu, secilen sayinin yerine konularak tum sayi havuzundan secilmesidir. 
  )
  sum(zarat)# Burada sum ile toplanarak bir geri donus verisi olusturuldu. 
}           
            # Bu asamada Fonksiyon olusturuldu. 

roll()      # Bu satirda Fonksiyon cagirilir.


roll2 <- function(bones){     # bones ile fonksiyonu dinamik hale getirir. 
  zarat <- sample(
    bones,
    size = 2,
    replace = TRUE
  )
  sum(bones)
}

roll2(3)        # Fonksiyona deger gondererek calismasini saglar.
```
